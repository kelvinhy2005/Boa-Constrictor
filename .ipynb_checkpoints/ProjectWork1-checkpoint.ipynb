```python
# Import Dependencies
import pandas as pd
import numpy as np
from config import api_key
import requests
from pprint import pprint
```

```python
# Define API base url and targetted years of movies data
url1 = "https://api.themoviedb.org/3/discover/movie?"
start_date = 2007
end_date = 2017
```

```python
# Call the API url , Check for response and print the response whether the data is available or not
# store the repsonse as JSON data
response = requests.get(f"{url1}api_key={api_key}&primary_release_year.gte={start_date}&primary_release_year.lte={end_date}")
movies_data = response.json()
pprint(movies_data)
movies_data_list = [movie for movie in movies_data["results"]]
pprint(movies_data_list)
```

```python
# Get movie id list from the original data pulled based on number of years
# This information is used to in the second API call --> To pull data based on years and do the analysis, we need two API calls

movie_id_list = []

for movie_id in movies_data_list:
    movie_id_list.append(movie_id["id"])

print(movie_id_list)   
```

```python
# Using movie id list call the TMDb API for more movie information
url2 = "https://api.themoviedb.org/3/movie/"
movie_details = []

for mve_id in movie_id_list:
    
    response2 = requests.get(f"{url2}{mve_id}?api_key={api_key}")
    movie_details.append(response2.json())
    
pprint(movie_details)
```

```python
movie_id1 = []
movie_title = []
budget = []
revenue = []
release_date = []
genre = []
production_comp = []
tag_line = []
overview = []

for detail in movie_details:
   # pprint(detail)
    movie_id1.append(detail["id"])
    movie_title.append(detail["original_title"])
    release_date.append(detail["release_date"])
    budget.append(detail["budget"])
    revenue.append(detail["revenue"])
    tag_line.append(detail["tagline"])
    overview.append(detail["overview"])
    genre.append(detail["genres"][0]["name"])
    #production_comp.append(detail["production_companies"]["name"])

movie_df_frame = pd.DataFrame({"ID":movie_id1,"Title": movie_title, "Budget":budget, "Revenue": revenue, "Genre": genre,"Tag Line": tag_line,"Overview":overview,"Release Date":release_date})
movie_df_frame.head()
    
```

```python
#Dropped rows with 0 revenus or budget
movie_df_frame = movie_df_frame[movie_df_frame.Budget!=0]
movie_df_frame = movie_df_frame[movie_df_frame.Revenue!=0]
movie_df_frame.head()
```

```python
movie_df_frame["ROI"] = round(((movie_df_frame["Revenue"]-movie_df_frame["Budget"])/movie_df_frame["Budget"])*100,2)

movie_df_frame.head()
```
