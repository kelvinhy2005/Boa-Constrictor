```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn as sns
import os
import csv

csv = os.path.join('santhosh.csv')
movie_df = pd.read_csv(csv,encoding='latin-1')
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
movie_df['Tagline'].astype(str)
movie_df = movie_df[movie_df.Budget!=0]
movie_df = movie_df[movie_df.Revenue!=0]
movie_df.head()
```

```python
Compound = []
comp=[]

for tagline in movie_df['Tagline']:
    comp = analyzer.polarity_scores(tagline)["compound"]
    Compound.append(comp)
movie_df['Sentiment Score']=Compound
movie_df['ROI'] = round((movie_df['Revenue']-movie_df['Budget'])/movie_df['Budget'],2)

movie_df.head()
```

```python
movie_df = movie_df.dropna()
movie_df['Sentiment Score'].astype(int)

movie_df['ROI'].astype(int)
print(movie_df['ROI'])
```

```python
import seaborn as sns
sns.jointplot('ROI','Sentiment Score',data=movie_df,kind="reg",xlim=[0,60],ylim=[-1,1])
```
